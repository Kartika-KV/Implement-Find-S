```python
import csv

#all given attributes of the training data set 

attributes=[['Sunny','Rainy,','Winter'],
           ['Warm','Cold'],
            ['Normal','High'],
           ['Strong','Weak'],
           ['Warm','Cold'],
           ['Same','Change']]
len_a=len(attributes)
print("No of Attributes are ",len_a)
print("The most General Hypothesis is ['?','?','?','?','?','?']")
print("The most Specific Hypothesis is ['0','0','0','0','0','0']")

#creating empty list to insert our training data for evaluation 
training_set=[]
print("The Training Data given is ")
with open("ws.csv") as train:
    reader=csv.reader(train)
    for i in reader:
        training_set.append(i)
        print(i)
hypothesis=[0]*(len_a)
print("The intial Hypothesis is ",hypothesis)
for k in range(0,len_a):
    hypothesis[k]=training_set[0][k]
print(hypothesis)
"""for j in range(0,len_a):
    print(hypothesis[j],end=" ")"""

#Finding a general Hypothesis from a given specific hypothesis 
for i in range(0,len(training_set)):
    print(training_set[i][len_a])
    if training_set[i][len_a]=='Yes':
        for j in range(0,len_a):
            if training_set[i][j]==hypothesis[j]:
                hypothesis[j]=training_set[i][j]
            else:
                hypothesis[j]='?'
    print("Hypothesis is ",hypothesis)
        
   
        

print(hypothesis)
```
